```python
def score(corr_miss_info, df_user, name_miss_info, df_user_number):        #df_user: ligne correspondant à l'utilisateur
    #corr_miss_info: colonne de corrélation de l'information à estimer
    score = 0
    for column in corr_miss_info.columns.tolist()[:61]:
        if df_user.at[df_user_number,column] != np.nan:
            score += df_user.at[df_user_number,column]*corr_miss_info.at[name_miss_info,column]
    for column in corr_miss_info.columns.tolist()[84:]:
        if df_user.at[df_user_number,column] != np.nan:
            score += df_user.at[df_user_number,column]*corr_miss_info.at[name_miss_info,column]
    return score

def estimation_zipcode(corr_df, df_user_number, dataframe):     #corr_df matrice (type: dataframe) de corrélation du dataframe 
    df_user = dataframe.iloc[[df_user_number]].copy()               #dataframe: dataframe des données
    scores = []
    for i in range(24): 
        corr_area = corr_df.iloc[[60 + i]]                      #corr_area: renvoie le dataframe de corrélation correspondant à la région 65+i
        area_name = corr_df.index.tolist()[60 + i]
        score1 = score(corr_area, df_user, area_name, df_user_number)
        scores += [score1]
    m = max(scores)
    k = 0
    while scores[k] != m:
        k += 1
    return corr_df.index.tolist()[60 + k]
```

```python
from random import randint
j = 0
k = 0
for i in range(100):
    n = randint(1,100000)
    estimate = estimation_zipcode(corr_matrix,n,data)
    if estimate == "['region']_ILE-DE-FRANCE":
        k += 1
    elif estimate == "['region']_MIDI-PYRENEES":
        j += 1
    print(estimate,data.loc[n].tolist()[61:83]) #,estimation_zipcode2(M,n,data_file,list_population)])
print("Ile de France:",k,"Midi-Pyrénnées:",j)
    
```
